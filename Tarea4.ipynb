```python
import time
import matplotlib.pyplot as plt

# Calculo de pi utilizando fórmula más rápida (con su respectivo error)
def pi_rapido(error_relativo):
    pi_aprox = 3
    n = 1
    denom = 2
    while True:
        termino = 4 / (denom * (denom + 1) * (denom + 2))
        if n % 2 == 1:
            pi_aprox += termino
        else:
            pi_aprox -= termino
        n += 1
        denom += 2
        error = abs(pi_aprox - 3.14159265358979323) / 3.14159265358979323
        if error < error_relativo:
            break
    return pi_aprox

# Calculo  pi usando fórmula de Leibniz (con su respectivo error)
def pi_leibniz(error_relativo):
    pi_aprox = 0
    n = 0
    denom = 1
    while True:
        termino = 4 / denom
        if n % 2 == 0:
            pi_aprox += termino
        else:
            pi_aprox -= termino
        n += 1
        denom += 2

        error = abs(pi_aprox - 3.14159265358979323) / 3.14159265358979323
        if error < error_relativo:
            break
    return pi_aprox


# Listas que almacenan los tiempos de ejecución de cada algoritmo y lista que contiene valores de error relativo que se usaran para calcular
tiempos_leibniz = []
tiempos_rapido = []
valores_errores_relativos = [0.1, 0.095, 0.083, 0.076, 0.068, 0.054, 0.047, 0.033, 0.025, 0.01]

# medida de los tiempos de ejecucion de ambos algoritmos
for error_relativo in valores_errores_relativos:
    inicio_tiempo = time.time()
    pi_leibniz(error_relativo)
    fin_tiempo = time.time()
    tiempos_leibniz.append(fin_tiempo - inicio_tiempo)

    inicio_tiempo = time.time()
    pi_rapido(error_relativo)
    fin_tiempo = time.time()
    tiempos_rapido.append(fin_tiempo - inicio_tiempo)

# Graficos de los tiempos de ejecución vs errores relativos
plt.plot(errores_relativos, tiempos_leibniz, label='Algoritmo de Leibniz')
plt.plot(errores_relativos, tiempos_rapido, label='Algoritmo más rápida')
plt.xlabel('Error Relativo')
plt.ylabel('Tiempo de Ejecución (segundos)')
plt.title('Tiempo de ejecución vs. error relativo')
plt.show()


```

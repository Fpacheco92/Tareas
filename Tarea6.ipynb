```python
import numpy as np
import matplotlib.pyplot as plt

#se genera una distribucion normal de 100 datos que arroja la distribucion discreta de poisson en 10 segundos con rapidez esperada de 0.99c y error asociado de 0.01c

Dist_NormVelocidades = np.random.normal(0.99, 0.01, np.random.poisson(100/6,1))

#histograma de los datos generados
plt.hist(Dist_NormVelocidades)
plt.title("distribucion de velocidades")
plt.ylabel("NÂ° de eventos")
plt.xlabel("velocidad obtenida")
plt.show()
```

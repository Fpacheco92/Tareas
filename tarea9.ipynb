```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Funcion a integrar con la variable random asociada
def p(x):
    return np.sin(x) + np.random.normal(0, 1e-5)

# valor real de la integral
valor_real = 1 - np.cos(1)

# metodo de cuadratura adaptativa
def cuad_adaptativa(func, a, b, tol=1e-6):
    resultado, error = quad(func, a, b, epsabs=tol)
    return resultado, error

# Calculo integral p(x) en intervalo (0, 1) bajo distintas tolerancias
a, b = 0, 1
tolerancias = [1e-2, 1e-4, 1e-6, 1e-8, 1e-10, 1e-12, 1e-14]  # distintos rangos de tolerancia 
errores = []

for tol in tolerancias:
    resultado, error = cuad_adaptativa(p, a, b, tol)
    errores.append(np.abs(resultado - valor_real))

# grafico pedido
plt.figure(figsize=(8, 6))
plt.loglog(tolerancias, errores, marker='o', linestyle='-', color='b')
plt.title('Error vs Tolerancia')
plt.xlabel('Tolerancia')
plt.ylabel('Error Absoluto')
plt.show()



# Analisis de estabilidad
if errors[-1] / errors[-2] < 10:
    print("El metodo de cuadratura adaptativa es estable.")
else:
    print("El metodo de cuadratura adaptativa puede presentar inestabilidad.")


# Es posible ver en el grafico que para determinados valores de tolerancia es posible alcanzar
# la precision deseada, no se puede especificar un intervalo ya que la funcion depende de un parametro random
#cambia cada vez q se compila.







```
